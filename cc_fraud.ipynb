# Split Dataset for test & train

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x_smote, y_smote, test_size=0.20, random_state=42)
```

# Model Building

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, Y_train)
```

# Generating predictions

Predict transactions as fraudulent or genuine by giving random sample

```python
Y_pred_logistic = logistic_model.predict(X_test)
```

Classification report

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
classification_report_logistic = classification_report(Y_test, Y_pred_logistic)

print(classification_report_logistic)
```

Confusion matrix

```python
confusion_matrix_logistic = confusion_matrix(Y_test, Y_pred_logistic)

confusion_matrix_logistic
```

```python
confusion_matrix_df = pd.DataFrame(confusion_matrix_logistic,
                                   columns=["Predicted Negative (0)", "Predicted Positive (1)"],
                                   index=["Actual Negative (0)", "Actual Positive (1)"])

confusion_matrix_df
```

*   fpr (false positive rate) = evaluate the performance of the model, particularly in the context of binary classification tasks

*   thresholds = adjust the sensitivity and specificity of a classifier by changing the balance between false positives and false negatives

Calculate ROC-AUC score

```python
from sklearn.metrics import roc_curve, auc
y_scores = logistic_model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(Y_test, y_scores)
roc_auc = auc(fpr, tpr)
```

Plot ROC curve

```python
plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
